```python
path1 = 'E:/Phonepe_Pulse_project/Data/data/aggregated/transaction/country/india/state'
agg_trans_lis = os.listdir(path1)

columns = {
    "State": [],
    "Year": [],
    "Quarter": [],
    "Transaction_type": [],
    "Transaction_count": [],
    "Transaction_amount": []
}

for state in agg_trans_lis:
    cur_state = os.path.join(path1, state)
    agg_year_lis = os.listdir(cur_state)

    for year in agg_year_lis:
        cur_year = os.path.join(cur_state, year)
        agg_file_lis = os.listdir(cur_year)

        for file in agg_file_lis:
            cur_file = os.path.join(cur_year, file)
            with open(cur_file, "r") as data:
                a = json.load(data)

            for i in a['data']['transactionData']:
                name = i['name']
                count = i['paymentInstruments'][0]['count']
                amount = i["paymentInstruments"][0]["amount"]
                columns["Transaction_type"].append(name)
                columns["Transaction_count"].append(count)
                columns["Transaction_amount"].append(amount)
                columns["State"].append(state)
                columns["Year"].append(year)
                columns["Quarter"].append(int(file.strip(".json")))

df_agg_trans = pd.DataFrame(columns)
df_agg_trans.head()
```

```python
path2='E:/phonepey/pulse-master/data/aggregated/user/country/india/state'
agg_user_lis = os.listdir(path2)

columns1 = {
    "State": [],
    "Year": [],
    "Quarter": [],
    "Brands": [],
    "Count": [],
    "Percentage": []
}

for state in agg_user_lis:
    cur_state = os.path.join(path2, state)
    agg_year_lis = os.listdir(cur_state)

    for year in agg_year_lis:
        cur_year = os.path.join(cur_state, year)
        agg_file_lis = os.listdir(cur_year)

        for file in agg_file_lis:
            cur_file = os.path.join(cur_year, file)
            with open(cur_file, "r") as data:
                b = json.load(data)
                try:
                    for i in b["data"]["usersByDevice"]:
                        brand_name = i["brand"]
                        counts = i["count"]
                        percents = i["percentage"]
                        columns1["Brands"].append(brand_name)
                        columns1["Count"].append(counts)
                        columns1["Percentage"].append(percents)
                        columns1["State"].append(state)
                        columns1["Year"].append(year)
                        columns1["Quarter"].append(int(file.strip('.json')))
                except:
                    pass
df_agg_user = pd.DataFrame(columns1)
```

```python
path3 = 'E:/phonepey/pulse-master/data/map/transaction/hover/country/india/state'

map_tarns_lis=os.listdir(path3)

columns2 = {
    "State": [],
    "Year": [],
    "Quarter": [],
    "District": [],
    "Count": [],
    "Amount": []
}

for state in map_tarns_lis:
    cur_state = os.path.join(path3,state)
    map_year_lis = os.listdir(cur_state)

    for year in map_year_lis:
        cur_year = os.path.join(cur_state , year)
        map_file_lis=os.listdir(cur_year)
       
        for file in map_file_lis:
            cur_file = os.path.join(cur_year,file)
            with open(cur_file , "r") as data :
                c = json.load(data)
                
                for i in c['data']['hoverDataList']:
                    district = i['name']
                    count = i['metric'][0]['count']
                    amount = i['metric'][0]['amount']
                    columns2['District'].append(district)
                    columns2["Count"].append(count)
                    columns2["Amount"].append(amount)
                    columns2['State'].append(state)
                    columns2['Year'].append(year)
                    columns2["Quarter"].append(int(file.strip('.json')))
                    
df_map_trans = pd.DataFrame(columns2)
```

```python
path4 = "E:/phonepey/pulse-master/data/map/user/hover/country/india/state"

map_user_lis = os.listdir(path4)

columns3 = {
    "State": [],
    "Year": [],
    "Quarter": [],
    "District": [],
    "RegisteredUser": [],
    "AppOpens": []
}

for state in map_user_lis:
    cur_state = os.path.join(path4 ,state)
    map_year_lis = os.listdir(cur_state)
    
    for year in map_year_lis:
        cur_year = os.path.join(cur_state , year)
        map_file_lis = os.listdir(cur_year)
        

        for file in map_file_lis :
            cur_file = os.path.join(cur_year,file)
            with open(cur_file , 'r') as data:
                d = json.load(data)

                for i in d["data"]["hoverData"].items():
                    district = i[0]
                    registereduser = i[1]["registeredUsers"]
                    appOpens = i[1]['appOpens']
                    columns3["District"].append(district)
                    columns3["RegisteredUser"].append(registereduser)
                    columns3["AppOpens"].append(appOpens)
                    columns3['State'].append(state)
                    columns3['Year'].append(year)
                    columns3['Quarter'].append(int(file.strip('.json')))
                    
df_map_user = pd.DataFrame(columns3)
```

```python
path5 = 'E:/phonepey/pulse-master/data/top/transaction/country/india/state/'

top_trans_lis = os.listdir(path5)

columns4 = {
    "State": [],
    "Year": [],
    "Quarter": [],
    'Pincode': [],
    'Transaction_count': [],
    'Transaction_amount': []
}

for state in top_trans_lis:
    cur_state = os.path.join(path5 ,state)
    map_year_lis = os.listdir(cur_state)
    
    for year in map_year_lis:
        cur_year = os.path.join(cur_state , year)
        map_file_lis = os.listdir(cur_year)

        for file in map_file_lis:
            cur_file = os.path.join(cur_year, file)
            with open(cur_file , 'r') as data:
                e = json.load(data)

                for i in e['data']['pincodes']:
                    name = i['entityName']
                    count = i['metric']['count']
                    amount = i['metric']['amount']
                    columns4['Pincode'].append(name)
                    columns4['Transaction_count'].append(count)
                    columns4['Transaction_amount'].append(amount)
                    columns4['State'].append(state)
                    columns4['Year'].append(year)
                    columns4['Quarter'].append(int(file.strip('.json')))

df_top_trans=pd.DataFrame(columns4)
```
